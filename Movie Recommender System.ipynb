```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text): # to convert the list of string into list
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert) # we are getting the genre for all the movies
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords']
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['crew'][0] # director

# too many dictionaries, we'll just see the dictionary where -> job = director
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.info()
```

```python
movies['overview'].apply(lambda x : x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
# we'll remove spaces in between any two words to avoid confusion-
# e.h. there are 2 director Akshat Agrawal and Akshat Goyal - tags will be created for them, but lately model may get confused as
# the tags for Akshat will be same

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
# create tags


movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
# convert tags columns which has list values into strings by joining them with a " "

new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0] # combination of overview, genre, keywords, cast, crew
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower()) # converting tags to lowercase
```

```python
new.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()



def stemming(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ". join(y)



new['tags'] = new['tags'].apply(stemming)
new.head()
```

```python
# we'll convert tags into vectors and find similarity between vectors and taht's ow we find similar movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')  # considering 5000 most frequent words
```

```python
cv.get_feature_names() # 5000 most frequent words
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector
```

```python
vector[1] # vector representation of 1st movie
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
len(similarity[1])
```

```python
# we calculate the cosine similarity of each vector with all other vector

# In higher dimension Euclidean distance is not a good measure

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity.shape # har movie ka har movie k sath distance
```

```python
similarity[0] # first movie ka har(4806) movie k sath distance
```

```python
similarity[0].shape
```

```python
# now the approach is 
# jo b movie user btaega uska index dkhege, and then us index ka similarity matrix me se distance ko descending order me sort 
#krege and the movies which have max similarity will be recommended
```

```python
# to fetch index
index = new[new['title'] == 'Batman Begins'].index[0]
index

# this movie is at index 119
```

```python
sorted(similarity[0], reverse=True) # through this we'll get sorted list in descending order(most similar movies at the top)

# jo movies jad similar h wo upar aa gyi, but by doing this we loose the indexes. kaise pta chlega ki kon c movie ki baat ho rhi h.
# so we use enumerate
```

```python
list(enumerate(similarity[0]))
# telling each movie's index and it's similarity with all other movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True)

# sorting is done on the basis of index, not values

# so we use key = x[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])

```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
# key = lambda x: x[1] - telling to do sorting based on cosine similarity values not on the basis of index, agr hm ye use ni krege
# to sorting will be done on the basis of index not values
# reverse = True means descending order
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
recommend('Green Lantern')
```
